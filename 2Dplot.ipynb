### Barplot

```python
xLabel = ['X1', 'X2', 'X3', 'X4']
# yVal = random.sample(numbers, 4)
yVal = np.random.choice(intNumbers, size = 4, replace = False)
print("y Values:", yVal)

colors = sns.color_palette("Pastel1", 4)

plt.figure(figsize = (5,5))
plt.bar(np.arange(0,4), yVal, tick_label = xLabel, color = colors)

plt.title('Barplot', pad = 15, fontsize = 20)

plt.ylabel('Y Values', labelpad = 10, fontsize = 15)
plt.xlabel('X Values', labelpad = 10, fontsize = 15)

# "XLabel"
plt.xticks(fontsize = 10, rotation = 90)
# "yVal"
plt.yticks(fontsize = 10)
```

```python
cases1 = [1,3,5]
cases2 = [2,4,6]

cases = [f"{case1}*{case2}" for case1 in cases1 for case2 in cases2]

xLabel = ['X1', 'X2', 'X3']
yArray = np.random.randint(1,101, size = (len(cases), len(xLabel)))

colors = sns.color_palette("Pastel1", len(xLabel))

# Define the width of each bar
bar_width = 1.5
# Define how much you want to skip
skip = 8
# Define the positions on the x-axis
x = np.arange(len(cases))

for i in range(len(cases)):
    for j in range(len(xLabel)):
        plt.bar(skip * i + bar_width * j, yArray[i][j], width = bar_width, label = xLabel[j], color = colors[j]) 

# Set xticks
plt.xticks(skip * x + bar_width, cases, rotation = 0)
plt.xlim(-skip, skip*len(cases) + skip)

# Set legend
plt.legend(labels = xLabel, fontsize = 15, loc = 'upper left', bbox_to_anchor = (1, 1))

print(yArray)
```

### Boxplot

```python
# Using plt
xLabel = ['X1', 'X2', 'X3']
yArray = np.array([np.random.uniform(0,100, size = 50), 
                   np.random.uniform(0,50, size = 50), 
                   np.random.uniform(0,75, size = 50)])

colors = sns.color_palette("Pastel1", len(xLabel))

plt.figure(figsize = (5,5))
box = plt.boxplot(yArray.T, showmeans = True,labels = xLabel, patch_artist=True)

for patch, median, mean, color in zip(box['boxes'],box['medians'], box['means'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.5)
    median.set_linewidth(3)
    mean.set_marker('o')
    mean.set_markerfacecolor('white')
    mean.set_markeredgecolor('black')

# print(yArray)
# print(yArray.T)
```

#### Stripplot with Statistical Analysis

```python
from scipy import stats

xLabel = ['X1', 'X2', 'X3', 'X4', 'X5']
yArray = np.array([np.random.uniform(0, 100, size=50), 
                   np.random.uniform(0, 50, size=50), 
                   np.random.uniform(0, 75, size=50),
                   np.random.uniform(0, 50, size=50),
                   np.random.uniform(0, 80, size=50)])

dfData = pd.DataFrame(yArray.T, columns=xLabel)

colors = sns.color_palette("Pastel1", len(xLabel))

ax = sns.stripplot(data=dfData, palette=colors, size=5)
ax1 = sns.pointplot(data=dfData, estimator='mean', errorbar='sd', join=False, markers='o', 
                    scale=0.7, errwidth=0.5, capsize=0.2, color='black')

# Perform t-tests between pairs of columns
pairs = [('X1', 'X2'), ('X2', 'X3'), ('X3', 'X4'), ('X4', 'X5')]

# Define significance levels and stars
def get_star_annotation(pVal):
    if pVal < 0.0001:
        return '****'
    elif pVal < 0.001:
        return '***'
    elif pVal < 0.01:
        return '**'
    elif pVal < 0.05:
        return '*'
    else:
        return 'ns'  # Not significant

# Annotate the plot with significance stars
for (x1, x2) in pairs:
    t_stat, pVal = stats.ttest_ind(dfData[x1], dfData[x2])
    
    # Get the appropriate star annotation
    annotation = get_star_annotation(pVal)
    
    # Determine where to place the annotation on the plot
    yMax = dfData[[x1, x2]].max().max()  # Get the max y-value for the two groups being compared
    yPos = yMax + 5
    height = 3

    # Plot the significance lines and annotation
    x1Pos = xLabel.index(x1)
    x2Pos = xLabel.index(x2)
    ax.plot([x1Pos, x1Pos, x2Pos, x2Pos], [yPos, yPos + height, yPos + height, yPos], lw=1.5, c='black')
    ax.text((x1Pos + x2Pos) / 2, yPos + h, annotation, ha='center', va='bottom', color='black')

```

```python
from statannot import add_stat_annotation

xLabel = ['X1', 'X2', 'X3', 'X4', 'X5']
yArray = np.array([np.random.uniform(0, 100, size=50), 
                   np.random.uniform(0, 50, size=50), 
                   np.random.uniform(0, 75, size=50),
                   np.random.uniform(0, 50, size=50),
                   np.random.uniform(0, 80, size=50)])

dfData = pd.DataFrame(yArray.T, columns=xLabel)

colors = sns.color_palette("Pastel1", len(xLabel))

ax = sns.stripplot(data=dfData, palette=colors, size=5)
ax1 = sns.pointplot(data=dfData, estimator='mean', errorbar='sd', join=False, markers='o', 
                    scale=0.7, errwidth=0.5, capsize=0.2, color='black')

# Define the pairs of comparisons for statistical testing
pairs = [('X1', 'X2'), ('X2', 'X3'), ('X3', 'X4'), ('X4', 'X5'), ('X2', 'X4')]

# Add statistical annotation with t-tests and star format
add_stat_annotation(ax, data=dfData, x=None, y=None,
                    box_pairs=pairs,
                    test='t-test_ind', text_format='star',
                    loc='inside', verbose=2)
```

### Violinplot

```python
xLabel = ['X1', 'X2', 'X3', 'X4', 'X5']
yArray = np.array([np.random.uniform(0,100, size = 50), 
                   np.random.uniform(0,50, size = 50), 
                   np.random.uniform(0,75, size = 50),
                   np.random.uniform(0,50, size = 50),
                   np.random.uniform(0,80, size = 50)])

dfData = pd.DataFrame(yArray.T, columns = xLabel)

colors = sns.color_palette("Pastel1", len(xLabel))

# Draw violinplot
ax = sns.violinplot(data = dfData, palette = colors, linewidth = 0.5, inner = 'point')

for violin in ax.collections:
    violin.set_alpha(0.6)
```
