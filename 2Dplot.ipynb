## Plot

### Barplot

```python
xLabel = ['X1', 'X2', 'X3', 'X4']
# yVal = random.sample(numbers, 4)
yVal = np.random.choice(intNumbers, size = 4, replace = False)
print("y Values:", yVal)

colors = sns.color_palette("Pastel1", 4)

plt.figure(figsize = (5,5))
plt.bar(np.arange(0,4), yVal, tick_label = xLabel, color = colors)

plt.title('Barplot', pad = 15, fontsize = 20)

plt.ylabel('Y Values', labelpad = 10, fontsize = 15)
plt.xlabel('X Values', labelpad = 10, fontsize = 15)

# "XLabel"
plt.xticks(fontsize = 10, rotation = 90)
# "yVal"
plt.yticks(fontsize = 10)
```

```python
cases1 = [1,3,5]
cases2 = [2,4,6]

cases = [f"{case1}*{case2}" for case1 in cases1 for case2 in cases2]

xLabel = ['X1', 'X2', 'X3']
yArray = np.random.randint(1,101, size = (len(cases), len(xLabel)))

colors = sns.color_palette("Pastel1", len(xLabel))

# Define the width of each bar
bar_width = 1.5
# Define how much you want to skip
skip = 8
# Define the positions on the x-axis
x = np.arange(len(cases))

for i in range(len(cases)):
    for j in range(len(xLabel)):
        plt.bar(skip * i + bar_width * j, yArray[i][j], width = bar_width, label = xLabel[j], color = colors[j]) 

# Set xticks
plt.xticks(skip * x + bar_width, cases, rotation = 0)
plt.xlim(-skip, skip*len(cases) + skip)

# Set legend
plt.legend(labels = xLabel, fontsize = 15, loc = 'upper left', bbox_to_anchor = (1, 1))

print(yArray)
```

```python
cases1 = [1,3,5]
cases2 = [2,4,6]

cases = [f"{case1}*{case2}" for case1 in cases1 for case2 in cases2]

xLabel = ['X1', 'X2', 'X3']
yArray = np.random.randint(1,101, size = (len(cases), len(xLabel)))

colors = sns.color_palette("Pastel1", len(xLabel))

# Define the width of each bar
bar_width = 1.5
# Define how much you want to skip
skip = 8
# Define the positions on the y-axis 
y = np.arange(len(cases))

for i in range(len(cases)):
    for j in range(len(xLabel)):
        plt.barh(skip * i + bar_width * j, yArray[i][j], height = bar_width, label = xLabel[j], color = colors[j]) 

# Set yticks
plt.yticks(skip * y + bar_width, cases, rotation = 0)
plt.ylim(-skip, skip*len(cases) + skip)

# Set legend
plt.legend(labels = xLabel, fontsize = 15, loc = 'upper left', bbox_to_anchor = (1, 1))

print(yArray)
```

```python
cases1 = [1,3,5]
cases2 = [2,4,6]

cases = [f"{case1}*{case2}" for case1 in cases1 for case2 in cases2]

xLabel = ['X1', 'X2', 'X3']
yArray = np.random.randint(1,101, size = (len(cases), len(xLabel)))

colors = sns.color_palette("Pastel1", len(xLabel))

# Define the width of each bar
bar_width = 0.5
# Define the positions on the x-axis
x = np.arange(len(cases))


# Plot bars
plt.bar(x, yArray[:,0], width = bar_width, color = colors[0], label = xLabel[0])
plt.bar(x, yArray[:,1], bottom = yArray[:,0], width = bar_width, color = colors[1], label = xLabel[1])
plt.bar(x, yArray[:,2], bottom = yArray[:,0] + yArray[:,1], width = bar_width, color = colors[2], label = xLabel[2])

# Set xticks
plt.xticks(x, cases, rotation = 0)

# Set legend
plt.legend(labels = xLabel, fontsize = 15, loc = 'upper left', bbox_to_anchor = (1, 1))

print(yArray)
```

### Boxplot

```python
# Using plt
xLabel = ['X1', 'X2', 'X3']
yArray = np.array([np.random.uniform(0,100, size = 50), 
                   np.random.uniform(0,50, size = 50), 
                   np.random.uniform(0,75, size = 50)])

colors = sns.color_palette("Pastel1", len(xLabel))

plt.figure(figsize = (5,5))
box = plt.boxplot(yArray.T, showmeans = True,labels = xLabel, patch_artist=True)

for patch, median, mean, color in zip(box['boxes'],box['medians'], box['means'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.5)
    median.set_linewidth(3)
    mean.set_marker('o')
    mean.set_markerfacecolor('white')
    mean.set_markeredgecolor('black')

# print(yArray)
# print(yArray.T)
```

```python
# Using sns
xLabel = ['X1', 'X2', 'X3']
yArray = np.array([np.random.uniform(0,100, size = 50), 
                   np.random.uniform(0,50, size = 50), 
                   np.random.uniform(0,75, size = 50)])

dfData = pd.DataFrame(yArray.T, columns = xLabel)

colors = sns.color_palette("Pastel1", len(xLabel))

# Draw boxplot
ax = sns.boxplot(data = dfData, order = xLabel, palette = colors, width = 0.3, 
                 boxprops = dict(alpha = 0.5),
                 medianprops = dict(color = 'orange', linewidth = 3))

# Draw means
ax1 = sns.pointplot(data = dfData, estimator = np.mean, join = False, markers = 'o',
                    color = 'white', errorbar = ('sd'), errwidth=0.1)
```

### Heatmap

```python
arr = np.array(fltNumbers).reshape(10,10)

ax = sns.heatmap(arr, vmin = -1, vmax = 1, annot = False, cmap = 'coolwarm')

# Ticks of the heatmap
ax.tick_params(left = False, bottom = False, labelleft = False, labelbottom = False)

# Colorbar of the heatmap
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(size = 3, labelsize = 12)
```
